```sql
%python
import boto3
import pandas as pd
import random
import time
from datetime import datetime

# AWS Credentials (Same as before)
ACCESS_KEY = "AKIA6IWCNKFVPKSUFJIM"
SECRET_KEY = "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"
BUCKET_NAME = "upi-fraud-data-project-2025"

s3 = boto3.client('s3', aws_access_key_id=ACCESS_KEY, aws_secret_access_key=SECRET_KEY)

def generate_live_data(n=5):
    states = ["Maharashtra", "Delhi", "Karnataka", "Telangana", "Uttar Pradesh"]
    banks = ["SBI", "ICICI", "HDFC", "Axis", "PNB"]
    
    data = []
    for _ in range(n):
        now = datetime.now()
        txn = {
            "transaction id": f"TXN_LIVE_{random.randint(100000, 999999)}",
            "timestamp": now.strftime("%Y-%m-%d %H:%M:%S"),
            "transaction type": random.choice(["P2P", "P2M", "Bill Payment"]),
            "merchant_category": random.choice(["Grocery", "Food", "Entertainment", "Shopping"]),
            "amount (INR)": round(random.uniform(10, 50000), 2),
            "transaction_status": "SUCCESS",
            "sender_age_group": random.choice(["18-25", "26-35", "36-45"]),
            "receiver_age_group": random.choice(["18-25", "26-35", "36-45"]),
            "sender_state": random.choice(states),
            "sender_bank": random.choice(banks),
            "receiver_bank": random.choice(banks),
            "device_type": random.choice(["Android", "iOS"]),
            "network_type": random.choice(["4G", "5G", "WiFi"]),
            "fraud_flag": 1 if random.random() > 0.95 else 0, # 5% chance of fraud
            "hour_of_day": now.hour,
            "day_of_week": now.strftime("%A"),
            "is_weekend": 1 if now.weekday() >= 5 else 0
        }
        data.append(txn)
    return pd.DataFrame(data)

# Infinite loop to simulate "Live" flow
print("Starting Live Data Stream to S3...")
while True:
    df_live = generate_live_data()
    file_name = f"live_data/txn_{int(time.time())}.csv"
    
    # Convert to CSV and upload to S3
    csv_buffer = df_live.to_csv(index=False)
    s3.put_object(Bucket=BUCKET_NAME, Key=file_name, Body=csv_buffer)
    
    print(f"Uploaded {file_name} to S3")
    time.sleep(10) # Wait 10 seconds before next batch
```
